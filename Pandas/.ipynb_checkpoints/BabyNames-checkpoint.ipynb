```python
names_all = []

for year in range(1880,2014+1):
    names_all.append(pd.read_csv('names/yob{}.txt'.format(year),names=['name','sex','number']))
    names_all[-1]['year'] = year #adds the column year to the dataframe and assigns the value
    
allyears = pd.concat(names_all) #concatenate dataframes inside list of dataframes
```

```python
def plotname(sex,name):
    data = allyears_indexed.loc[sex,name]
    
    pp.plot(data.index,data.values)
```

```python
def topten(sex,year):
    simple = allyears_indexed.loc[sex,:,year].sort_values('number',ascending=False).reset_index()
    simple = simple.drop(['sex','year','number'],axis=1).head(10)
    
    simple.columns = [year]
    simple.index = simple.index + 1
    
    return simple
```

```python
topten('M',2007)
```

```python
def toptens(sex,year0,year1):
    years = [topten(sex,year) for year in range(year0,year1+1)]
    
    return years[0].join(years[1:])
```

```python
toptens('F',1900,2014).stack().value_counts()
```

```python
popular = toptens('F',1985,1995).stack().value_counts().index[:6]
```

```python
pp.figure(figsize=(12,2.5))
for name in popular:
    plotname('F',name)
pp.legend(popular)
```

```python
allyears.groupby(['sex','name']).sum().head()
```

```python
def sumsq(freq):
    return sum(freq**2)

totals = allyears.groupby(['sex','name'])['number'].agg('sum')
```

```python
spikyness = allyears.groupby(['sex','name'])['number'].agg(sumsq) / totals**2
#spikyness is a number between 0 and 1. 1 happens when a name appears only once in a year
spikyness.head()
```

```python
spiky_common = spikyness[totals > 5000].copy()
spiky_common.sort_values(ascending=False).head(10)
```

```python
spiky_common.sort_values(ascending=False).tail(5)
```

```python
pp.figure(figsize=(12,2.5))

plotname('F','Louisa')
plotname('M','Shaquille')
```

```python
fads = spiky_common.sort_values(ascending=False).head(10).index.values
```

```python
pp.figure(figsize=(12,2.5))

for sex,name in fads:
    plotname(sex,name)
    
pp.legend([name for sex,name in fads], loc='upper left')
```

```python
totals_recent = allyears[allyears['year']>2005].groupby(['sex','name'])['number'].sum()
```

```python
#spikyness = allyears[allyears['year']>2005].groupby(['sex','name'])['number'].agg(sumsq) / totals_recent**2
spiky_common = spikyness[(totals > 5000) & (totals_recent < 1000)].copy()
spiky_common.sort_values(ascending=False).head(10)
```

```python
fads = spiky_common.sort_values(ascending=False).head(10).index.values
```

```python
pp.figure(figsize=(12,2.5))

for sex,name in fads:
    plotname(sex,name)
    
pp.legend([name for sex,name in fads], loc='upper left')
```

```python
allyears.head()
names_total = allyears.groupby(['sex','name'])['number'].sum()

totals_bysex = names_total.unstack('sex')
totals_bysex.head()
```

```python
totals_both = totals_bysex.sum(axis=1)
totals_both.head()
```

```python
ratio = totals_bysex['F'] / totals_bysex['M']
ratio.head()
```

```python
unisex = (ratio>0.25) & (ratio<4)
```

```python
totals_both = totals_both[unisex]
totals_both.sort_values(ascending=False).head(10)
```

```python
names = totals_both.sort_values(ascending=False).head(10).index.values
names
```

```python
pp.figure(figsize=(12,12))

for i,name in enumerate(names):
    pp.subplot(5,2,i+1)
    
    plotname('M',name)
    plotname('F',name)
    
    pp.legend([name + ' M', name + ' F'],loc='upper left')
```
