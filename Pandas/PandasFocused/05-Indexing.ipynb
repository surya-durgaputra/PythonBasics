### Indexing
- The index object is an immutable array, and indexing allows you to access a row or a column using a label.
- This is what makes Pandas special, because typically in other programming languages, you cannot access an array using labels.m

```python
import pandas as pd
```

```python
oo = pd.read_csv('data/olympics.csv', skiprows=4)
oo.head()
```

```python
# lets look at the type of Index
type(oo.index)
```

- we see that index has its own type and is not series or a dataframe.

```python
# this return the index of the 100th element of oo, which in this case is 100
oo.index[100]
```

### Indexes are immutable, i.e. we cannot change them

```python
# this will result in an error as indexes are immutable
oo.index[100] = 5
```

### set_index()
- The set index allows us to determine which of the series is going to be the index.
- We will use the set index with one key parameter, which is the keys.
- We will look at inplace in particular
- set_index(keys,inplace=True)

#### thre are two ways of doing set_index(), inplace and otherwise

```python
oo.head()
```

```python
# here you see the default index is the series of 0,1,2,3...,
# we can change this like so:
oo.set_index('Athlete').head()
```

```python
# make these changes inplace,
oo.set_index('Athlete',inplace=True)
oo.head()
# now oo itself has changed
```

```python
# we can reset index
oo.reset_index(inplace=True)
oo.head()
```

```python
ath = oo.set_index('Athlete')
ath.head()
```

### reset_index()
 - the opposite of setting an index is to reset the index
 - what we're doing here is we're returning a DataFrame to its default, integer-based index. 

```python
# lets reset the ath dataframe so that it is again indexed by integers
ath.reset_index(inplace=True)
ath.head()
```

#### In the real world, you can often use the reset index, especially when you're trying to determine which is the best series to use as an index for your data frame.

### sort_index()
- Sort index allows for all the items to be sorted by that index.
- The advantage of this is that when you have a particularly large data set, sorting the index reduces the time to access any subset of that data.
- You can sort objects by a label along the axis.

```python
ath.head()
```

```python
ath.set_index('Athlete',inplace=True)
ath.head()
```

```python
ath.sort_index(inplace=True)
ath.head()
```

```python
#to change the sorting order to descending
ath.sort_index(inplace=True, ascending=False)
ath.head()
# you would be surprised to see ÖSTRAND than a name starting with Z at the end of ascending order.
# This is because with this encoding, the standard alphabet will have a lower value than the special 
# characters used in other languages.
# So if your data has special characters, you need to take this into account.
```

### loc[]
- loc[] is a label-based indexer, that means you are selecting by the labels.
    - DataFrame.loc[]
    - DataFrame.Series.loc[]
- And notice that "loc[]" uses square brackets and not regular brackets.
- loc[] will raise the KeyError when items are not found.

```python
ath.sort_index(ascending=True,inplace=True)
ath.reset_index(inplace=True)
ath.head()
```

#### now let's try to find USAIN, Bolt
#### We have two ways to do this:
 - traditional way

```python
ath[ath.Athlete=='BOLT, Usain']
```

- using set_index() and loc[]

```python
ath.set_index('Athlete',inplace=True)
```

## NOTE: Running set_index() with the same key multiple times will result in error. It wil run first time when the key is not an index. If the key is already an index, it will return an error

```python
ath.loc['BOLT, Usain']
```

### iloc[]
- With iloc[], we're doing selection by integer index.
- Iloc is primarily integer position based (from 0 to length-1 of the axis).
- IMPORTANT: iloc[] is gerally used with default integer based index
- One of the advantages of iloc[] is that it allows for the traditional Pythonic slicing.

```python
ath.reset_index(inplace=True)
```

```python
ath.head()
```

```python
ath.iloc[1700]
#this will return the athlete at integer-index 1700
```

#### I can also use a list with iloc[], giving a list of integers

```python
ath.iloc[[1700,2000,13000]]
#this will return the rows corresponding to these indexes
```

#### since iloc[] allows Pythonic slicing

```python
ath.iloc[0:5]
# will return first five rows
```

### In the real world, I would often use one of the other series as my index, and so, I would typically use Loc for indexing.
### However, if my data is sorted in a way that allows me to use the default index, then I might use Iloc, especially when I want to access chunks of data using Python slicing.
