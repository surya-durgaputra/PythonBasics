```python
import pandas as pd
import os
import requests
import time
import re
import matplotlib.pyplot as plt
from datetime import datetime
from bs4 import BeautifulSoup

path = "C:/Users/DevAccessa/Documents/VS/Pt/intraQuarter/intraQuarter"
inx_file = "C:/Users/DevAccessa/source/repos/LoyaltyPrimePt/LoyaltyPrimePt/inx.csv"
#path = "C:\\Users\\DevAccessa\\Documents\\VS\\Pt\\intraQuarter"

def Key_Stats(gather=""):
    ticker_list = []
    statspath = path+'/_KeyStats'
    company_list_dirs = [x[0] for x in os.walk(statspath)]
    df = pd.DataFrame(columns=['Date','Unix','Ticker','DE Ratio','Price','price_p_value','SP500','sp500_p_value'])
    df_inx = pd.read_csv(inx_file)
    columns = df_inx.columns.values
    columns[0] = 'Date'
    df_inx.columns = columns
    df_inx.set_index('Date')
    for subdir in company_list_dirs[1:]:
        ticker = subdir.split('\\')[1]
        ticker_list.append(ticker)
        file_list = os.listdir(subdir)
        sp500_starting_value = False
        price_starting_value = False
        if len(file_list) > 0:
            for file in file_list:
                filepath = subdir+'\\'+file
                #filepath = 'C:/Users/DevAccessa/Documents/VS/Pt/intraQuarter/intraQuarter/_KeyStats/bac/20090203170742.html'
                f = open(filepath,'r')
                soup = BeautifulSoup(f,'html.parser')
                f.close()

                price = None
                tables = soup.find_all('table',{'class':'yfnc_leftnav1'})
                if len(tables) > 0:
                    for table in tables:
                        big = table.find('big')
                        if big is not None:
                            b = big.find('b')
                            if b is not None:
                                price = b.string
                                break
                    #price = table.find('big').find('b').string
                if price is None:
                    time_rtq_ticker = soup.find('span',{'class':'time_rtq_ticker'})
                    if time_rtq_ticker is not None:
                        span = time_rtq_ticker.find('span')
                        if span is not None:
                            price = span.string
                    
                date_stamp = datetime.strptime(file,'%Y%m%d%H%M%S.html')
                unix_time = time.mktime(date_stamp.timetuple())
                matches = soup.find_all('td',string=re.compile(gather))
                if len(matches)>0:
                    for match in matches:
                        try:#in some cases 'DE Ratio':float(data.string) will cause exception since value cant be converted to float
                            try:
                                sp500_date = datetime.fromtimestamp(unix_time).strftime('%Y-%m-%d')
                                df_inx.set_index('Date')
                                row_sp500 = df_inx[(df_inx.Date==sp500_date)]
                                sp500_value = float(row_sp500['adjusted close'])
                            except Exception as e:
                                sp500_date = datetime.fromtimestamp(unix_time-259200).strftime('%Y-%m-%d')
                                row_sp500 = df_inx[(df_inx.Date==sp500_date)]
                                sp500_value = float(row_sp500['adjusted close'])
                            
                            if sp500_starting_value == False:
                                sp500_starting_value = sp500_value
                            if price_starting_value == False:
                                price_starting_value = float(price)
                            data = match.find_next_sibling('td')
                            price_p_value = ((float(price)-price_starting_value)/price_starting_value)*100
                            sp500_p_value = ((sp500_value-sp500_starting_value)/sp500_starting_value)*100
                            line = {'Date':str(datetime.fromtimestamp(unix_time).strftime('%Y-%m')),
                                    'Unix':unix_time,
                                    'Ticker':ticker,
                                    'DE Ratio':float(data.string),
                                    'Price':float(price),
                                    #we calculate the differential over time
                                    'price_p_value':price_p_value,
                                    'SP500':sp500_value,
                                    'sp500_p_value':sp500_p_value,
                                    'Difference': price_p_value - sp500_p_value}
                            #try:
                            #    sp500_date = datetime.fromtimestamp(unix_time).strftime('%d/%m/%Y')
                            df = df.append(line,ignore_index=True)
                        except Exception as e:
                            pass
    #for ticker in ticker_list:
    #    try:
    #        plot_df = df[(df.Ticker==ticker)]
    #        plot_df = plot_df.set_index('Date')
    #        plot_df.Difference.plot(label=ticker)
    #    except:
    #        pass
    #plt.xticks(df.index, df.Date, rotation=90)
    #plt.legend()
    #plt.show()
    save = gather.replace(' ','').replace('/','').replace('\\','').replace('(','').replace(')','')+('.csv')
    df.to_csv(save)
#Key_Stats(r'Total Debt/Equity \(mrq\)')


def Plot():
    df = pd.read_csv('C:/Users/DevAccessa/source/repos/LoyaltyPrimePt/LoyaltyPrimePt/TotalDebtEquitymrq.csv')
    tickers = list(set(df.Ticker))
    tickers = sorted(tickers)
    try:
        for ticker in tickers:
            plot_df = df[(df.Ticker == ticker)]
            print('date length:',len(list(plot_df['Date'])))
            plot_df = plot_df.set_index('Date')
            print(plot_df.head())
            print('ticker:',ticker)
            
            plot_df.Difference.plot(label=ticker)
            break
        plt.xticks(df.index,df.Date,rotation=90)
        plt.legend()
        plt.show()
    except:
        pass

Plot()

#df = pd.read_csv('TotalDebtEquitymrq.csv')
#print(df['Unix'].idxmax())
#print(df.iloc[df['Unix'].idxmax()])
#print('-------------------')
#print(df['Unix'].idxmin())
#print(df.iloc[df['Unix'].idxmin()])




#import pandas_datareader.data as web
#https://pydata.github.io/pandas-datareader/devel/remote_data.html#remote-data-alphavantage

#df = web.DataReader("INX","av-daily-adjusted",start=datetime(2003,1,1),end=datetime(2013,12,12),access_key='0G8JUDEZ8P8IILKL')
#df.to_csv('inx.csv')
```
