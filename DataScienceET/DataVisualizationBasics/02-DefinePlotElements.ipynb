### Defining axes, ticks, and grids .. Object oriented approach

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()

ax = fig.add_axes([.1, .1, 1, 1])
# [.1, .1, 1, 1] - in the order of appearance
# .1 = 10% of width of figure. This is the origin for x-axis. No effect if only one plot in fig.
# .1 = 10% of height of figure. This is the origin for y-axis. No effect if only one plot in fig.
# 1 = 100% of available width
# 1 = 100% of available height
ax.plot(x,y)
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()

ax = fig.add_axes([.1, .1, 1, 1])
# [.1, .1, 1, 1] - in the order of appearance
# .1 = 10% of width of figure. This is the origin for x-axis. 
# .1 = 10% of height of figure. This is the origin for y-axis
# 1 = 100% of available width
# 1 = 100% of available height
ax1 = fig.add_axes([.85, .9, .2, .1]) # add another plot in the same figure
# [.85, .9, .2, .1] - in the order of appearance
# .85 = 85% of width of figure. This is the origin for x-axis. 
# .9 = 90% of height of figure. This is the origin for y-axis
# .2 = 20% of available width
# .1 = 10% of available height
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1]) 

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1]) 

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x, y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

#Creates just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
#Creates two subplots and unpacks the output array immediately
f, (ax1,ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
#Creates four polar axes, and accesses them through the returned array
fig, axes = plt.subplots(2, 2, subplot_kw=dict(polar=True))
axes[0, 0].plot(x, y)
axes[1, 1].scatter(x, y)
```
