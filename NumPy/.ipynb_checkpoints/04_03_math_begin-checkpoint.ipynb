```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,40)
x
```

```python
sinx = np.sin(x)#numpy has its own way for math functions. 
#We don't use the standard math library for many things like this one 
sinx
```

```python
pp.plot(x, sinx)
```

```python
cosx = np.cos(x)
cosx
```

```python
# we can do arithmetical opeartion between arrays
y = sinx * cosx 
z = cosx**2 - sinx**2 #more complex arithmetic operation
pp.plot(x,y)
pp.plot(x,z)
pp.legend(['sin(x) cos(x)','cos(x)^2-sin(x)^2'])
```

```python
pp.plot(x,sinx,'x')
pp.plot(x,cosx,'o')
```

# Normally mathematical operations are applied to  arrays element by element. However if you want to do linear algebra, that is not the case

## for example, if we want to take the inner product of two vectors i.e. the sum of the element by element products

```python
np.dot(sinx,cosx)
```

## we can take the outer product which builds every possible combination from the two vectors. The result is a matrix.

```python
mat_outer = np.outer(sinx, cosx)
mat_outer
```

# Numpy always tries to be helpful in any way to guess what you want to do.
# So it has some broadcasting rules, with which it will try to make sense of operations between arrays of different shapes.
# For instance, if you have a one-dimensional vector, and you add a number to it, just a single number, the number will be added to every element.

```python
v = np.linspace(0,10,5)
v
```

```python
v+1
```

# Broadcasting also applies if we try to add a one-dimensional array to a two-dimensional array.
## If I add a one-dimensional array of size n to a two-dimensional array of size n-by-n. The result is a two-dimensional array where the one-dimensional array has been added to every row.

```python
vv = np.outer(v,v)
vv # a two dim array
```

```python
vv + v # add a 1-dim array to a 2-dim array
```

## If instead we wanted to add it to every column, we'd first have to turn it into proper n-by-one column vector, by adding a dimension with numpy newaxis.

```python
v_col = v[:,np.newaxis]
v_col # we have converted a 1xn array to nx1 array
```

```python
vv + v_col
```
